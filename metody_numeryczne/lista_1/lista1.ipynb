```python
import matplotlib.pyplot as plt
import numpy as np
import timeit
```

# Zadanie 1

```python
#definicja funkcji Padego
def z1(x):
    return (6-2*x)/(6+4*x+np.power(x,2))

def z2(x):
    return (6-4*x+np.power(x,2))/(6+2*x)
```

```python
x = np.linspace(-2,20,1000)
```

```python
#wykresy przybliżeń razem
plt.plot(x,np.exp(-x), label="exp(-x)",color = "blue", linewidth = 2)
plt.plot(x,z1(x), label="Padego z1",color="orange")
plt.plot(x,z2(x), label="Padego z2", color="brown")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
plt.title("Porównanie funkcji exp(-x) z przybliżeniami Padego")
plt.savefig("assets/porownanie1.png", format='png', dpi=300)
```

```python
#wykresy przybliżeń osobno
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(x,np.exp(-x),label="exp(-x)", color="blue",linewidth = 2)
ax1.plot(x,z1(x),label="Padego z1", color="orange")
ax1.legend()
ax1.set_xlabel("x")
ax1.set_ylabel("y")
ax1.set_title("Padego z1")

ax2.plot(x,np.exp(-x),label="exp(-x)", color = "blue", linewidth = 2)
ax2.plot(x,z2(x),label="Padego z2", color="brown")
ax2.legend()
ax2.set_xlabel("x")
ax2.set_ylabel("y")
ax2.set_title("Padego z2")

fig.suptitle("Porównanie funckji exp(-x) z przybliżeniami Padego z osobna", fontsize=16)

plt.tight_layout()  
plt.savefig("assets/porownanie2.png", format='png', dpi=300)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(x,abs(z1(x)-np.exp(-x)),color = "orange")
ax1.set_title("błąd bezwględny przybliżenia z1")
ax1.set_xlabel("x")
ax1.set_ylabel("błąd")

ax2.plot(x,z2(x)-np.exp(-x),color = "brown")
ax2.set_xlabel("x")
ax2.set_ylabel("błąd")
ax2.set_title("błąd bezwględny przybliżenia z2")
plt.tight_layout()  
plt.savefig("assets/bledy1.png", format='png', dpi=300)
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(10, 5))

ax1.plot(x,abs(z1(x)-np.exp(-x))/np.exp(-x),color = "orange")
ax1.set_title("błąd wględny przybliżenia z1")
ax1.set_xlabel("x")
ax1.set_ylabel("błąd")

ax2.plot(x,abs(z2(x)-np.exp(-x))/np.exp(-x),color = "brown")
ax2.set_xlabel("x")
ax2.set_ylabel("błąd")
ax2.set_title("błąd wględny przybliżenia z2")
plt.tight_layout()  
plt.savefig("assets/bledy2.png", format='png', dpi=300)
plt.show()
```

```python
values_range = np.arange(-10,10.0001,0.0001)
def w(x):
    return 6*np.power(x,4) + 5*np.power(x,3) - 13*np.power(x,2) + x + 1

def time_function():
    w(values_range)

print("Czas wykonania to: {}".format(timeit.timeit(time_function,number=1)))
```
