```python
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
import scipy.stats as sps
```

```python
def msr(x,y):
    #zwraca próbkowe beta_1 i beta_0
    n = len(x)
    a = (n * np.sum(x*y) - np.sum(x) * np.sum(y) ) / (n * np.sum(x**2) - np.sum(x)**2)
    b = (np.sum(x**2) * np.sum(y) - np.sum(x) * np.sum(x*y)) / (n*np.sum(x**2) - np.sum(x)**2)
    return [a,b]
```

```python
def msr_2(x,y):
    product = np.zeros(len(x))
    for i in range(len(x)):
        product[i] = x[i]*y[i]
    return np.sum(product)/np.sum(x**2)
        
```

## różne wielkości n

```python
n= np.arange(50,1050,50)
beta_0 = 5
beta_1 = 2
m = 5000
sigma = 2

probkowa_srednia_beta_1 = np.zeros(len(n))
probkowa_wariancja_beta_1 = np.zeros(len(n))
teoretyczna_wariancja_beta_1 = np.zeros(len(n))
teoretyczna_srednia_beta_1 = beta_1

# ==============================
for j in range(len(n)) :
    x = np.linspace(0,10,n[j])
    beta_1_hat_vector = np.zeros(m)
    for i in range(m):
        epsilons = np.random.normal(0,sigma,n[j])
        y = beta_1*x + epsilons
        beta_1_hat_vector[i] = msr_2(x,y)

    probkowa_srednia_beta_1[j] = np.mean(beta_1_hat_vector)
    probkowa_wariancja_beta_1[j] = np.var(beta_1_hat_vector)
    teoretyczna_wariancja_beta_1[j] = (sigma**2)/np.sum(np.power(x,2)) 
```

```python
def get_b1_b0_estims(x, sigma, b_0, b_1):
    n = len(x)
    epsilon = np.random.normal(0,sigma,n)
    y = b_0 + b_1*x + epsilon
    return msr(x,y,n)
```

```python
n = 1000
b_0 = 2
b_1 = 5
sigma = 2
x = np.linspace(0,10,n)
m = 1000
```

```python
b_0_hat = np.zeros(m)
b_1_hat = np.zeros(m)

for i in range(m):
    b_1_hat[i], b_0_hat[i] = get_b1_b0_estims(x, sigma, b_0, b_1)

var_theo_b_1 = (sigma**2)*(1/np.sum((x-np.mean(x))**2))
var_theo_b_0 = sigma**2*((1/n) + np.mean(x)**2/np.sum((x-np.mean(x))**2))
```

```python
def find_err(k, y_vals, x_vals):
    x1, y1 = x_vals[:k], y_vals[:k] 
    x2, y2 = x_vals[k:], y_vals[k:]

    b1_1, b0_1 = msr(x1,y1) 
    b1_2, b0_2 = msr(x2,y2)

    error1 = 0
    for i in range(len(x1)):
        pred = b0_1 + b1_1*x1[i]
        error1 += (y1[i]-pred)**2

    error2 = 0
    for i in range(len(x2)):
        pred = b0_2 + b1_2*x2[i]
        error2 += (y2[i]-pred)**2
    
    return error1 + error2

def k_point(X, C):
    n = len(X)
    errors = np.zeros(n-4)

    for k in range(2, n-2):
        errors[k-2] = find_err(k,C,X)
    
    optimal_split = np.argmin(errors) + 2
    return optimal_split
```

```python
#wyznaczenie x
n = 1000
L = 400
sigma1 = 1
sigma2 = np.sqrt(5)

rng = np.arange(1, n+1)

part1 = np.random.normal(0, sigma1, L)
part2 = np.random.normal(0, sigma2, n-L)

X = np.zeros(n)
X[:L] = part1
X[L:] = part2

C = np.cumsum( X**2 )
found_split = n-k_point(X,C)
```

```python
plt.plot(rng,X)
plt.axvline(L, color="red", label = "L")
plt.axvline(found_split, color="orange", label = "found L")
print(found_split)
plt.legend()
```
